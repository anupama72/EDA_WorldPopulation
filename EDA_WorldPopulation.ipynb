```python
#Looking for correlation between values
df.corr(numeric_only= True)
```

```python
#visualizing correlation with a heatmap using seaborn and matplotlib
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.rcParams['figure.figsize'] = (15,6)
plt.show()
```

```python
#are there certain countries that are growing faster than others
df2 = df.groupby('Continent')[df.columns[-5:-13:-1]].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
df2
```

```python
#transposing dataframe
df3 = df2.transpose()
df3
```

```python
#visualizing through a plot 
df3.plot()

#can decipher asia grew the fastest whereas oceania is almost a straight line
```
